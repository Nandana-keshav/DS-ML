```python
import matplotlib.pyplot as plt
data1=[72,76,24,40,57,62,75,78,31,32]
data2=[62,5,91,25,36,32,96,95,30,90]
data3=[23,89,12,78,72,89,25,69,68,86]

bdata=[data1,data2,data3]
box=plt.boxplot(bdata,vert=1,patch_artist=True)

colors=['pink','yellow','violet']
for p, color in zip(box['boxes'],colors):
    p.set_facecolor(color)
plt.show()
```

```python
import matplotlib.pyplot as plt

unemployment = [6.1, 5.8, 5.7, 5.7, 5.8, 5.6, 5.5, 5.3, 5.2, 5.2]
stockin = [1500, 1520, 1525, 1523, 1515, 1540, 1545, 1560, 1555, 1565]

plt.scatter(unemployment, stockin, c='b', marker='*')
plt.xlabel('Unemployment Rate ', fontsize=16)
plt.ylabel('Stock Index Value', fontsize=16)
plt.title('Unemployment vs Stock Index', fontsize=18)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x ** 2  
plt.plot(x, y)
plt.xlabel('x', fontsize=14)
plt.ylabel('y = x²', fontsize=14)
plt.title('Quadratic Function: y = x²', fontsize=16)
plt.show()
```
